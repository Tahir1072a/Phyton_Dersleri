```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting Data / Verilerin çizilmesi
Verileri bir grafikte göstermek için scatter() fonksiyonunu kullanabiliriz. Scatter fonksiyonu bir x dizisi ve bir y dizisi olmak üzere temel olarak iki parametre alır. Ekstra olarak da marker ve c(color) parametreleri ile grafik üzerinde gösterilen marker(işaretçi) çeşidi ve markerin rengini ayarlayabiliriz.

```python
x_train = np.array([1,2])
y_train = np.array([300,500])

plt.scatter(x_train, y_train, marker="x", c="red")
plt.title("Ev Fiyatları")
plt.ylabel("Price (1000x dolar)")
plt.xlabel("Price (1000x m^2)")
plt.show()
```

Burada modelin davranışını simule etmeye çalışıyoruz. Modelimiz aslında burada yaptığımız şeyi yapmaya çalışıyor. Weight(ağırlık) ve bias(esneklik değeri) üzerinde oynama yaparak doğru fonksiyonu üretmeye çalışıyor. w = 100 ve b = 100 için modelimiz doğru çalışmayacaktır. Ancak w = 200 ve b = 100 için modelimizi eğitim verisi ile %100 uyuşan bir fonksiyon ürettiğini anlar ve artık tahmin için bu fonksiyonu kullanır.
> **Note**: Aşağıdaki fonksiyon bizim girdiğimiz parametrelerie göre bir tahmin dizisi döndürür.

> **Note**: np.zeros(m) fonksiyonu parametre olarak aldığı uzunluk kadar bir dizi oluşturur ve bunun içini 0 değerleri ile doldurur.

```python
w = 200
b = 100

def compute_model_output(x, w, b):
    """
    
    :param x: Tek boyutlu bir numpy array 
    :param ww: Model parametresi
    :param bb: Model parametresi
    :return: f_wb(ndarray) model preduction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b # f_wb = wx + b
    
    return f_wb
```

**plt.plot()**: Plot fonksiyonu parametere olarak aldığı değerlerin geçtiği bir çizgi grafiği çizer. Ayrıca 2 değer dizisi alırsa, plot(x,y) şeklinde birbirleri ile eşleme yapar ve y değerleri için bir çizgi çizer. x[i] değerine y[i] değeri karşılık gelecek şekilde bir grafik ortaya çıkar.

```python
prediction_f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, prediction_f_wb, c="b", label="Our prediction") # Verileri bir çizgi ile birbirine bağlar.
plt.scatter(x_train, y_train, marker="x", c="red", label="Actual Variables")

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend() # Bu fonksiyon grafiğin köşesinde labelların çıkmasını sağlar. 
plt.show()
```

Bir örnek ele alarak farklı şekilde kullanımları inceleyelim.

```python
x_train2 = np.array([0., 1, 2, 3, 4, 5])
y_train2 = np.array([0,  0, 0, 1, 1, 1])
```

> **Note**: x_train[x_train == 1] aslında x_train içindeki 1 olan değerleri getir demektir!
 
> **Note**: x_train[y_train == 1] ifadesi ise y_train içinde 1 olan değerlerin indexine göre bir true/false dizisi döner ve true değerlere karşılık gelen indexlerin değerlerini bize verir.

```python
pos = y_train2 == 1
neg = y_train2 == 0

fig,ax = plt.subplots(1,2,figsize=(8,3))
#plot 1, single variable
ax[0].scatter(x_train2[pos], y_train2[pos], marker='x', s=80, c = 'red', label="y=1")
ax[0].scatter(x_train2[neg], y_train2[neg], marker='o', s=100, label="y=0", facecolors='none', 
              edgecolors='blue',lw=3)

ax[0].set_ylim(-0.08,1.1)
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('one variable plot')
ax[0].legend()

plt.show()
```

Çizgisel galde de gösterim yapabiliriz. * ile göstermek içinde *- koymamız yeterlidir.

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g*-")
```
