```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
```

Çizim yapar. İlk girdiğimiz parametre x eksenini sonraki girdiğimiz parametre ise y eksenini belirtir. En son girilen parametre ise çizginin rengini belirtir.

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

Çizgisel galde de gösterim yapabiliriz. * ile göstermek içinde *- koymamız yeterlidir.

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g*-")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.1,0.1,0.5,0.5])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
myFigure = plt.figure()
eksen1 = myFigure.add_axes([0.1,0.1,1,1])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni Veri İsmi")
eksen1.set_ylabel("Y Ekseni Veri İsmi")
eksen1.set_title("Ana Grafik")

eksen1 = myFigure.add_axes([0.22,0.65,0.4,0.35])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni Veri İsmi")
eksen1.set_ylabel("Y Ekseni Veri İsmi")
eksen1.set_title("Küçük Grafik")
```

=> birinci parametre sol padding, 2. parametre down(alt) padding belirtir. 3. ve 4. parametre grafiğin boyurunu belirtir.
